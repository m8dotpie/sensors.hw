```python
import numpy as np
import matplotlib.pyplot as plt
```

# Task 1.1

## Generating Bimodal Data

```python
n_points = 1000
n_bins = 30
```

### First unimodal distribution

```python
np.random.seed(47)
n1 = np.random.normal(0, 1, n_points) + 4

fig, ax = plt.subplots(1, 1)
ax.hist(n1, bins=n_bins, density=True)
plt.plot()
```

### Second unimodal distribution

```python
np.random.seed(134)
n2 = np.random.normal(0, 1, n_points) / 2

fig, ax = plt.subplots(1, 1)
ax.hist(n2, bins=n_bins, density=True)
plt.plot()
```

### The bimodal distribution

```python
nt = np.concatenate((n1, n2))

fig, ax = plt.subplots(1, 1)
ax.hist(nt, bins=n_bins * 2, density=True)
plt.plot()
```

## Validating Distribution

### Small number of tests
Seemingly no pattern

```python
means = []
n_samples = 100
n_tests = 100

for i in range(n_tests):
    means.append(np.random.choice(nt, n_samples).mean())

fig, ax = plt.subplots(1, 1)
ax.hist(means, bins=n_bins * 2, density=True)
plt.plot()
```

### Large number of tests
Obvious pattern

```python
means = []
n_samples = 100
n_tests = 100000

for i in range(n_tests):
    means.append(np.random.choice(nt, n_samples).mean())

fig, ax = plt.subplots(1, 1)
ax.hist(means, bins=n_bins * 2, density=True)
plt.plot()
```

# Task 1.2

```python
# generating noisy polynomial data
np.random.seed(47)
n_points = 100
x = np.linspace(0, 1, n_points)
y = -2 * x**5 - x**4 + 6 * x**2 + 3 + np.random.normal(0, 0.2, n_points)

fig, ax = plt.subplots(1, 1)
ax.scatter(x, y)
plt.plot()
```

```python
# least squares regression
A = np.vstack([x**3, x**2, x, np.ones(len(x))]).T
m, c, b, a = np.linalg.lstsq(A, y, rcond=None)[0]

fig, ax = plt.subplots(1, 1)
ax.scatter(x, y)
ax.plot(x, a + b * x + c * x**2 + m * x**3, color='red')
plt.plot()
```
